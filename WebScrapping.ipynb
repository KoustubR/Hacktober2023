```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging

flipkart_url = 'https://www.flipkart.com/search?q='
search = 'tv'
desired_url = flipkart_url + search
urlclient = urlopen(desired_url)
urlclient
```

```python
fk_page = urlclient.read()
fk_html = bs(fk_page,  'html.parser')
fk_html
```

```python
fk_home = 'https://www.flipkart.com'
bigbox = fk_html.find_all('div' ,{'class': '_1AtVbE col-12-12'})
len(bigbox)
```

```python
del bigbox[0:3]
```

```python
fk_home + bigbox[0].div.div.div.a['href']
```

```python
list_url =[]
try:
    for i in bigbox:
        list_url.append(fk_home + i.div.div.div.a['href'])
except AttributeError as A:
    pass

list_url
```

```python
len(list_url)
```

```python
list_url_resp = {}
j=0
for i in list_url:
    list_url_resp[j]=[i , bs(requests.get(i).text, 'html.parser')]
    j+=1
    
list_url_resp[0]
```

```python
#comment_box = product.find_all('div', {'class' : '_16PBlm'})
for i in range(len(list_url_resp)): 
    try:
        list_url_resp[i] = [list_url_resp[i][0] , list_url_resp[i][1].find_all('div', {'class' : '_16PBlm'})]
    except AttributeError as A:
        pass
```
